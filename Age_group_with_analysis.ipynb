```python
vet_suc_age_list = (veterans_age_df['vet_suicide_2005']+veterans_age_df['vet_suicide_2006']+veterans_age_df['vet_suicide_2007']+veterans_age_df['vet_suicide_2008'])
vet_suc_pop = veterans_age_df['vet_pop_16sates']

nonvet_suc_age_list = (veterans_age_df['Nonvet_suicide_2005']+veterans_age_df['Nonvet_suicide_2006']+veterans_age_df['Nonvet_suicide_2007']+veterans_age_df['Nonvet_suicide_2008'])
nonvet_suc_pop = veterans_age_df['nonvet_pop_16sates']
```

```python
sum_total_vet_suicide = df4['Total_Vet_Sucide'].sum()
percent_vet_suicide = (df4['Total_Vet_Sucide']/sum_total_vet_suicide)*100
suicide_percentage = percent_vet_suicide.round(2)
```

# Among men, the risk was 19 percent higher among veterans compared to civilians.

```python
N = 7

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind,vet_rate, width, color='forestgreen', yerr=0.2*df3['Vet Std error'],capsize=7)


rects2 = ax.bar(ind+width, non_vet_rate, width, color='grey', yerr=0.2*df3['Non-Vet Std error'],capsize=7)

# add some text for labels, title and axes ticks
ax.set_ylabel('Suicide Rate per 100,000 Population')
ax.set_xlabel('Age Group')
ax.set_title('Suicide Rate by Age Group')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('18-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80+'))
plt.ylim(0,90,10)
ax.legend((rects1[0], rects2[0]), ('Veteran', 'Non-Veteran'))
sns.set(rc={'figure.figsize':(12,7)})

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.25*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```
