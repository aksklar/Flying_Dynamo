```python
file = "VHA_Center.csv"
file1 = 'veteran_suicides_2005-2011.csv'
file2 = '50_usstates_data.csv'
```

```python
vet_df = pd.read_csv(file1)
state_df = pd.read_csv(file2)
```

```python
vh_df = pd.read_csv(file,  encoding = "ISO-8859-1")

```

```python
merge_df = pd.merge(state_df, vh_df, how = 'outer', on='state abbreviation')
merge_df.shape
```

```python
vhacountdf = merge_df.groupby(['state']).count()
vhacount_df = vhacountdf.reset_index()
vhacount_df
```

```python
merge_df = pd.merge(vet_df, vhacount_df, how = 'outer', on='state')
merge_df.head()
```

```python
#deleting the unnamed column -cleaning 
del merge_df['Unnamed: 0']
merge_df.shape
```

```python
#Vet suicide Calculation 


merge_df1 = merge_df.set_index('state')
merge_df1.shape
```

```python
vet_suicideratelist = []
vetpopulationlist = []
statelist= []
state_c = []
vhcenterslist = []
```

```python
for index,row in merge_df1.iterrows():
    print(row[48])
    print(index)
    #Overall population from 2005 to 2011 satewise
    overallpop = (row[1] + row[7]+ row[13] + row[19] +row[25] + row[31] + row[37])
    vetpop = (row[0] + row[6] + row[12] + row[18] + row[24] + row[30] + row[36])
    
    vet_suicide = (row[3] + row[9]+ row[15]+ row[21]+row[27]+row[33]+row[39])
    
    vet_suiciderate = round((vet_suicide / vetpop) *100000, 2)
    facilityc = row[48]
    
    vetpopulationlist.append(vetpop)
    vet_suicideratelist.append(vet_suiciderate)
    statelist.append(index)
    vhcenterslist.append(facilityc)
    
print(vet_suicideratelist)
print(vetpopulationlist)
print(statelist)
print(len(vhcenterslist))
```

```python
# need have state abbrevation column for the map ploting 
state_df['state abbreviation']
```

```python
#list to dataframe
new_df = pd.DataFrame({'statelist':statelist, 'vet_suicideratelist':vet_suicideratelist, 'vetpopulationlist' : vetpopulationlist, 'vha centerslist': vhcenterslist, 'st_code' : state_df['state abbreviation']})
```

```python
#order new df 
new_df = new_df[['st_code','statelist', 'vet_suicideratelist','vetpopulationlist','vha centerslist']]
new_df
```

```python
for col in new_df.columns:
    new_df[col] = new_df[col].astype(str)
```

```python
new_df.columns
```

```python
scl = [[0.0, 'rgb(204, 255, 102)'],[0.4, 'rgb(179, 255, 26)'],\
            [0.6, 'rgb(136, 204, 0)'],[0.8, 'rgb(85, 128, 0)'],[1.0, 'rgb(0, 128, 0)']]
#[0.2, 'rgb(102, 255, 102)'],
```

```python
new_df['text'] = new_df['statelist'] + ':' + new_df['st_code'] + '<br>' +\
     'Veterans population ' + new_df['vetpopulationlist'] + '<br>' +\
     'Suiciderate ' + new_df['vet_suicideratelist'] + '<br>' + 'VHA Count ' + new_df['vha centerslist']
```

```python
data = [dict(
        type = 'choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = new_df['st_code'],
        z= new_df['vet_suicideratelist'],
        locationmode = 'USA-states',
        text = new_df['text'],
        marker = dict(
            line = dict(
                color = 'rgb(51,51,51)',
                width = 2
            )),
        colorbar = dict(
                title = "Veteran Suicide <br>State Comparison",
                titleside = 'left',
                tickmode = 'array',
                tickvals = [55,35,15],
                ticktext = ['High','Medium','Less'],
                ticks = 'outside'
                )
            )]
```

```python
layout = dict(
        title = 'Veteran Suicide State wise comparison',
        geo = dict(
            scope = 'usa',
            projection=dict(type = 'albers usa'),
            showlakes = True,
            lakecolor = 'rgb(0, 138, 230)'),
            )
        
```

```python
fig = dict(data= data, layout = layout)
choromap = go.Figure(data = data,layout = layout)
plotly.offline.plot(choromap)
```
