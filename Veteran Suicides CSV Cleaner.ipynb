```python
# TO RUN THIS SCRIPT:
# You must put the original files into a folder named "us-veteran-suicides-raw"
# AND you must created a folder named "us-veteran-suicides"

import pandas as pd
import csv
import numpy as np
```

```python
years = [2005, 2006, 2007, 2008, 2009, 2010, 2011]
for year in years:
    
    #Append each needed column's name with the year:
    csv_df = pd.read_csv(f"us-veteran-suicides-raw/{year}.csv")
    
    for x in np.arange(2,8,1):
        csv_df.columns.values[x] = csv_df.columns.values[x] + f"_{year}"
        
    csv_df.to_csv(f"us-veteran-suicides/{year}.csv")
    # Import a dataframe for each CSV
```
